## Build the quantum circuit

```python
# Build a quantum circuit
qc = QuantumCircuit(3,3) # 3 q-wires, 3 classical 

# Put X (NOT) gate on the 0-th wire
qc.x(0)  

# Put Hadamard gates on all wires
qc.h(range(3)) 

# Put a CNOT(0,1) gate 
qc.cx(0, 1) 

# Make measurements on all wires
#qc.measure(1,1) 
qc.measure(range(3),range(3)) 
```

## Print circuit

```python
# Print circuit
print(qc)
```

## Draw casually

```python
qc.draw()
```

## Draw with some glamor

```python
qc.draw(output='mpl')
```

## Do the same with classical and quantum registers

```python
# Define number of quantum regsiters
qr1 = QuantumRegister(3,'q')  

# Define number of classical regsiters
cr1 = ClassicalRegister(1,'c0')
cr2 = ClassicalRegister(2,'c1')


# Build a quantum circuit
qcnew = QuantumCircuit(qr1,cr1, cr2) # 3 q-wires, 3 classical 
#qcnew = QuantumCircuit(3,3)

# Put X (NOT) gate on the 0-th wire
qcnew.x(0)  

# Put Hadamard gates on all wires
qcnew.h(range(3)) 

# Put a CNOT(0,1) gate 
qcnew.cx(0, 1) 

# Make measurements on all wires
#qc.measure(1,1) 
qcnew.measure(range(3),range(3)) 
qcnew.draw('mpl')
```
